Since we're performing regression to predict median_house_value, let's split our data to features and target.

```python
X = df.drop('median_house_value',axis=1)
y = df['median_house_value']
```

```python
X.head()
```

```python
y.head()
```

```python
#Split your data into train and test
from sklearn.model_selection import train_test_split

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Remember that the features has some null values. So let's take care of that.

```python
x_train_df = pd.DataFrame(X_train, columns=X.columns)
x_test_df = pd.DataFrame(X_test, columns=X.columns)
```

```python
x_train_df.head()
```

```python
x_test_df.head()
```

```python
x_train_df.isnull().sum()
```

```python
x_test_df.isnull().sum()
x_test_df.columns.get_loc('total_bedrooms')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

fill_impute = SimpleImputer(strategy='constant', fill_value=df['total_bedrooms'].value_counts().idxmax())
sim = ColumnTransformer([('total_bedrooms',
                         fill_impute,
                         [4])])
x_test_df.iloc[:, [4]] = fill_impute.fit_transform(x_test_df.iloc[:, [4]])
```

```python
#Check if there's any null value left
x_test_df.isnull().sum()

```

```python
#Check for X_train as well
x_train_df.isnull().sum()
```

Now we're left with no null value. Good.

```python
#Encode the categorical features ['ocean_proximity'] to numerical value

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trans_column = ['ocean_proximity']
trans = ColumnTransformer(transformers=[('encode',
                                        OneHotEncoder(),
                                        trans_column)],
                         remainder='passthrough')
X_train = trans.fit_transform(x_train_df)
X_test = trans.transform(x_test_df)
```

Now our data is ready for fitting into an estimator

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import RandomizedSearchCV

np.random.seed(42)
random_dict = {'loss':['huber','epsilon_insensitive','squared_epsilon_insensitive'],
              'alpha':[0.01,0.001,0.0001]}

reg = SGDRegressor(max_iter=10000)
rs_reg = RandomizedSearchCV(estimator=reg,
                           param_distributions=random_dict,
                           n_iter=5,
                           cv=5)

rs_reg.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

np.random.seed(42)

y_pred = rs_reg.predict(X_test)

print('Regresssion model metrics on the test data')
print(f'R^2: {r2_score(y_test, y_pred)}')
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import cross_val_score

check = cross_val_score(rs_reg,
                       X_test,
                       y_test,
                       scoring='neg_mean_squared_error',
                       cv=5)
print('The mean absolute error is ')
print(np.sqrt(-check).mean())
```

```python
#Use another estimator

from sklearn.ensemble import RandomForestRegressor
for_reg = RandomForestRegressor(n_estimators=100)

for_reg.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

check = cross_val_score(for_reg,
                       X_test,
                       y_test,
                       scoring='neg_mean_squared_error',
                       cv=5)
print('The mean absolute error is ')
print(np.sqrt(-check).mean())
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

np.random.seed(42)

model = SVR()


grid_dict = {'kernel':['linear','poly','rbf'],
            'C':[1.0,10,100],
            'epsilon':[0.1,1.0,10]}

grid_model = GridSearchCV(model,grid_dict,cv=5)

grid_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

y_pred = grid_model.predict(X_test)

print('Regresssion model metrics on the test data')
print(f'R^2: {r2_score(y_test, y_pred)}')
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
```
